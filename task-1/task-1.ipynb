```python
# replacing 2ncycle with masters as both are same
df['Education'] = df['Education'].replace('2n Cycle','Master')
```

```python
# assigning numerical values to eduaction levels, higher the level, higher the value
df['Education'] = df['Education'].replace({'Basic':1,'Graduation':2,'Master':3,'PhD':4})
```

```python
df['Marital_Status'].unique()
```

```python
# reducing the number of categories by replacing same categories as single and married
df['Marital_Status'] = df['Marital_Status'].replace(['Single','YOLO','Absurd','Divorced','Widow','Alone'],'Single')
df['Marital_Status'] = df['Marital_Status'].replace(['Together','Married'],'Married')
```

```python
df['Marital_Status'].unique()
```

```python
df['Marital_Status'] = df['Marital_Status'].replace(['Single'],0)
df['Marital_Status'] = df['Marital_Status'].replace(['Married'],1)
```

```python
# converting date column from object tyoe to date-time type
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```
